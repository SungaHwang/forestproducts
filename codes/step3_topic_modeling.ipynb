```python
# 임산물 리스트
imsan_type = ['대추', '호두', '머루', '복분자', '산마늘', '산딸기', '죽순', '돌배', '고사리', '표고버섯',
            '도라지', '감', '눈개승마', '곤드레', '꽃송이버섯', '밤', '도토리', '고려엉겅퀴', '송이버섯',
            '더덕', '참나물', '어수리', '두릅', '석이버섯', '석류', '싸리버섯', '잣', '능이버섯',
            '목이버섯', '복령버섯', '취나물', '개암', '다래', '고비', '원추리', '은행', '곰취',
            '들메나무순']

# 임산물 단어를 제거하는 함수
def remove_imsan_type(text):
    for word in imsan_type:
        text = text.replace(word, "")
    return text

# '수정된제목' 컬럼의 각 문자열에 대해 단어 제거 적용
data['수정된제목'] = data['수정된제목'].apply(remove_imsan_type)
```

### BERTopic

```python
def preprocess_titles(titles):
    replacements = {
        "만드는법":"만들기",
        "만들기": "만들기",
        "만드는": "만들기",
        "만든":"만들기",
        "만들었어요": "만들기",
        "예약요":"예약",
        "예약받습니다":"예약",
        "판매합니다":"판매"

    }

    processed_titles = []
    for title in titles:
        for original, replacement in replacements.items():
            title = title.replace(original, replacement)
        processed_titles.append(title)

    return processed_titles
```

```python
from bertopic import BERTopic

# 제목 데이터 리스트
titles = preprocess_titles(data['수정된제목'])

from sklearn.feature_extraction.text import CountVectorizer

vectorizer_model = CountVectorizer(stop_words=["&quot",'&amp','떫은감','떫은','버섯','임산물','quot','amp'])

# 모델 초기화 및 학습
model = BERTopic(language="korean", nr_topics=15, vectorizer_model=vectorizer_model)
topics, probabilities = model.fit_transform(titles)

# 주제 출력
topics = model.get_topic_info()
topics
```

```python
# 각 제목이 어떤 topic에 속하는지 매칭하는 함수
def match_topic(title, topic5):
    for idx, row in topic5.iterrows():
        # 제목에 해당 topic의 Representation 리스트에서 하나라도 포함되면 매칭
        if any(keyword in title for keyword in row['Representation']):
            return row['topic']
    return '알 수 없음'  # 만약 매칭되지 않으면 '알 수 없음' 반환

# 'topic' 컬럼 생성하여 각 제목에 매칭된 topic을 추가
data['topic'] = data['정보획득사이트제목명'].apply(lambda x: match_topic(x, topic5))
```

### 예측값 컬럼 붙이기

```python
data_with_topic = pd.read_csv('/content/drive/MyDrive/[공모전] 임업통계 경진대회/codes/data/data_with_topic.csv')
```

```python
data_imsan_test = data_with_topic
```

```python
# 정보획득일자 -> 연도, 월 로 나누기
data_imsan_test['연도'] = data_imsan_test['정보획득일자'].astype(str).str[:4]
data_imsan_test['월'] = data_imsan_test['정보획득일자'].astype(str).str[4:6]

# NaN 값을 'false'로 채우기
data_imsan_test['이용자직업명'] = data_imsan_test['이용자직업명'].fillna('false')
data_imsan_test['가구소득그룹명'] = data_imsan_test['가구소득그룹명'].fillna('false')
data_imsan_test['주거형태명'] = data_imsan_test['주거형태명'].fillna('false')
data_imsan_test['자동차그룹명'] = data_imsan_test['자동차그룹명'].fillna('false')

# 결측치 및 불필요한 공백을 'false'로 채우기
data_imsan_test.loc[data_imsan_test['결혼여부명'] == '      ', '결혼여부명'] = 'false'
data_imsan_test.loc[data_imsan_test['자녀여부명'] == '          ', '자녀여부명'] = 'false'
data_imsan_test.loc[data_imsan_test['자동차그룹명'].str.strip() == '', '자동차그룹명'] = 'false'

# '자동차그룹명'의 연속된 공백 제거
data_imsan_test['자동차그룹명'] = data_imsan_test['자동차그룹명'].str.replace(r'\s{3,}', '', regex=True)

# 예측에 필요한 컬럼만 빼서 데이터프레임으로 만들기
test_data = data_imsan_test.copy()
test_data

# 카테고리형으로 모든 컬럼 변환
for col in test_data:
    test_data.loc[:, col] = test_data[col].astype('category')
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import os
import pickle

# 모델 경로 설정
model_dir = "/content/drive/MyDrive/[공모전] 임업통계 경진대회/codes/models"

# 각 임산물분류별로 지정된 모델
model_mapping = {
    '견과류': 'XGBoost',
    '버섯류': 'XGBoost',
    '나물류': 'XGBoost',
    '열매과일류': 'XGBoost',
    '뿌리줄기류': 'XGBoost'
}

# 결과 저장할 리스트 초기화
predictions = []

# 각 임산물분류별로 모델을 적용하여 예측 수행
for category, model_name in model_mapping.items():
    # 해당 임산물분류에 해당하는 데이터 필터링
    category_data = test_data[test_data['임산물분류'] == category]

    # 예측에 사용할 피처 선택
    X = category_data[['연도', '월', '이용자성별코드', '연령그룹코드', '이용자직업명', '가구소득그룹명', '결혼여부명',
                      '자녀여부명', '주거형태명', '이용자지역시도명', '이용자지역구명', '이용자지역동명']].copy()

    # X의 각 범주형 변수를 LabelEncoder로 인코딩
    label_encoders = {}
    for column in X.select_dtypes(include=['object']).columns:
        label_encoders[column] = LabelEncoder()
        X[column] = label_encoders[column].fit_transform(X[column])

    # 모델 파일 로드
    model_filename = os.path.join(model_dir, f"{category}_{model_name}_model.pkl")
    with open(model_filename, 'rb') as file:
        model = pickle.load(file)

    # 예측 수행
    y_pred = model.predict(X)

    # 예측 결과 저장
    category_predictions = category_data.copy()
    category_predictions['예측값'] = y_pred
    predictions.append(category_predictions)

# 예측 결과를 하나의 데이터프레임으로 결합
predictions_df = pd.concat(predictions, ignore_index=True)

# 예측 결과를 CSV 파일로 저장
predictions_csv_filename = "/content/drive/MyDrive/[공모전] 임업통계 경진대회/codes/data/result_step3.csv"
predictions_df.to_csv(predictions_csv_filename, index=False, encoding='utf-8-sig')
```
